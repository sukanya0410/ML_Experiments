```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Create Synthetic Dataset
np.random.seed(0)

# Generate 3 clusters of 2D points
cluster1 = np.random.normal([2, 2], 0.4, size=(150,2))
cluster2 = np.random.normal([7, 3], 0.5, size=(150,2))
cluster3 = np.random.normal([5, 8], 0.6, size=(150,2))

data = np.vstack((cluster1, cluster2, cluster3))

# Plot Original Data
plt.figure(figsize=(6,5))
plt.scatter(data[:,0], data[:,1], s=10)
plt.title("Original Data Points")
plt.show()

# Build Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, random_state=0)
gmm.fit(data)

# Predict Cluster Labels
labels = gmm.predict(data)

# Plot Clustered Output
plt.figure(figsize=(6,5))
plt.scatter(data[:,0], data[:,1], c=labels, cmap='viridis', s=10)
plt.title("Clusters after EM (GMM)")
plt.show()
```
